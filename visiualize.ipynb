```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

# 读取音频文件
audio_path = "/raid3/hm/workspace/freeau/high_freq/1725385273_steps_200_guidance_3.0_s1_2.5_s2_2.5_b1_1.0_b2_1.0_constant_tango2_seed27/AbplcXwXnvE_102751_30.wav"
y, sr = librosa.load(audio_path)

# 计算梅尔频谱
S = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=512, fmax=8192)

# 将梅尔频谱转换为对数刻度（dB）
S_dB = librosa.power_to_db(S, ref=np.max)

# 绘制梅尔频谱
plt.figure(figsize=(10, 4))
librosa.display.specshow(S_dB, sr=sr, x_axis='time', y_axis='mel', fmax=8192)
plt.axis('off')

# 保存频谱图为图像文件
plt.savefig('reference.png', dpi=300, bbox_inches='tight', pad_inches=0)
plt.show()
```
