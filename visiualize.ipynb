```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

# 
audio_path = "/raid2/DATA/ckpt/audiocaps_test/test/g6CY7qvu81k_105086_0.wav"  # 
y, sr = librosa.load(audio_path)

# 
S = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=512, fmax=8192)

# dB
S_dB = librosa.power_to_db(S, ref=np.max)

# 
plt.figure(figsize=(10, 4))
librosa.display.specshow(S_dB, sr=sr, x_axis='time', y_axis='mel', fmax=8192)
plt.axis('off')

# 
plt.savefig('reference.png', dpi=300, bbox_inches='tight', pad_inches=0)
plt.show()
```
